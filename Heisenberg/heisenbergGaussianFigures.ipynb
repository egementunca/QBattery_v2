Extracted Work Comparison

```python
deltaW = np.zeros(shape=(len(charger_names), realizations))
for i in range(len(charger_names)):
    for j in range(realizations):
        deltaW[i,j] = (ergs[i,j,-1])-min(ergs[i,j])
```

```python
plt.rcParams['axes.linewidth'] = 4
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

marker_list = ['o', 'v', '^', '<', '>', 's', 'D', 'P']

fig, axs = plt.subplots(1,1,figsize=(16,9), dpi=120)

for r in range(len(std_list)):
    axs.plot(np.arange(len(charger_names)), deltaW[::,r], marker=marker_list[r], markersize=20, linestyle='dashed', linewidth=3, label=r'$\sigma=$'+str(std_list[r]))
    axs.set_ylabel(r'$\Delta \mathcal{W}_{avg}$', fontsize=40)
    axs.set_xlabel('Configurations', fontsize=40)

axs.set_xticks(np.arange(len(charger_names)))
axs.set_xticklabels(charger_names, fontsize=40)

plt.legend(loc='lower right', framealpha=.5, fontsize=30)
plt.yticks(fontsize=40)

for e in ['x', 'y']:
    axs.tick_params(axis=e,which='major',size=16,width=2,direction='in',top='on',right='on')

plt.tight_layout()
plt.savefig('./Figures/heisenberg_gaussian_ancilla_erg_diff_.png')
plt.savefig('./Figures/heisenberg_gaussian_ancilla_erg_diff.eps', format='eps')
```

Our investigation into the behavior of the Heisenberg model under Gaussian bond strengths unveiled a rich interplay between standard deviation, ancillary spin numbers, and work extraction dynamics. Specifically, varying the standard deviation introduced intriguing changes in how the Heisenberg model responded to different configurations of ancillary spins.

When considering the influence of low standard deviations in the Gaussian bond strengths, a noteworthy pattern emerged. In scenarios where the standard deviation was low, we observed a direct and linear relationship between the number of ancillary spins and the extractable work. This linear trend indicates that, under these conditions, the battery's charging process is highly responsive to the number of ancillary spins, resulting in a proportional increase in extractable work as the number of spins grows.

Conversely, as we explored higher standard deviations in the Gaussian bond strengths, a distinctive behavior manifested. In these instances, the extractable work exhibited a convergence to a fixed value as the number of ancillary spins increased. This intriguing convergence phenomenon aligns with our findings from the fixed bond strengths scenario, where the battery system's work extraction stabilized into a consistent pattern regardless of the bond configurations.
