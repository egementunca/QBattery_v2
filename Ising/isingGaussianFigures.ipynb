Extracted Work Comparison

```python
deltaW = np.zeros(shape=(len(charger_names), realizations))
for i in range(len(charger_names)):
    for j in range(realizations):
        deltaW[i,j] = (ergs[i,j,-1])-min(ergs[i,j])
```

```python
plt.rcParams['axes.linewidth'] = 4
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

marker_list = ['o', 'v', '^', '<', '>', 's', 'D', 'P']

fig, axs = plt.subplots(1,1,figsize=(16,9), dpi=120)

for r in range(len(std_list)):
    axs.plot(np.arange(len(charger_names)), deltaW[::,r], marker=marker_list[r], markersize=20, linestyle='dashed', linewidth=3, label=r'$\sigma=$'+str(std_list[r]))
    axs.set_ylabel(r'$\Delta \mathcal{W}_{avg}$', fontsize=40)
    axs.set_xlabel('Configurations', fontsize=40)

axs.set_xticks(np.arange(len(charger_names)))
axs.set_xticklabels(charger_names, fontsize=40)

plt.legend(fontsize=30, framealpha=.5)
plt.yticks(fontsize=40)

for e in ['x', 'y']:
    axs.tick_params(axis=e,which='major',size=16,width=2,direction='in',top='on',right='on')

plt.tight_layout()

plt.savefig('./Figures/ising_gaussian_ancilla_erg_diff_.png')
plt.savefig('./Figures/ising_gaussian_ancilla_erg_diff.eps', format='eps')
```

When investigating the Ising model's behavior under different bond strengths, we observed a noteworthy connection between the bond strength distribution and the quantum battery's charging characteristics. Remarkably, the Ising model displayed behavior similar to that of plus or minus 1 bonds when the bond strengths were drawn from a Gaussian distribution.

Specifically, as the standard deviation of the Gaussian distribution increased, we noted a corresponding increase in ergotropy during the charging process. This finding indicates that higher bond strength variability introduces a cooperative element among ancillary spins. Thus, in the Ising model, higher bond strength variability translates into more efficient energy transfer, leading to enhanced extractable work.
