Extracted Work Comparison

```python
deltaW = np.zeros(shape=(len(charger_names), realizations))
for i in range(len(charger_names)):
    for j in range(realizations):
        deltaW[i,j] = (ergs[i,j,-1])-min(ergs[i,j])
```

```python
plt.rcParams['axes.linewidth'] = 4
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

marker_list = ['o', 'v', '^', '<', '>', ]

fig, axs = plt.subplots(1,1,figsize=(16,12), dpi=120)

x_axis = np.arange(len(charger_names))
for r in range(realizations):
    y_axis = np.array(deltaW[::,r])
    axs.plot(x_axis, y_axis, marker=marker_list[r], markersize=20,linestyle='dashed',linewidth=3, label=str(tuple(Jall[r])))
    axs.set_ylabel(r'$\Delta \mathcal{W}$', fontsize=60)
    axs.set_xlabel('Configurations', fontsize=60)


axs.set_xticks(np.arange(len(charger_names)))
axs.set_xticklabels(charger_names, fontsize=48)
plt.legend(fontsize=30)
plt.yticks(fontsize=48)

for e in ['x', 'y']:
    axs.tick_params(axis=e,which='major',size=16,width=2,direction='in',top='on',right='on')

plt.tight_layout()

plt.savefig('./Figures/ising_fixed_ancilla_erg_diff_43ratio.png')
plt.savefig('./Figures/ising_fixed_ancilla_erg_diff_43ratio.eps', format='eps')
```

Upon delving into the charging behavior of the Ising model, distinct characteristics emerged that highlight its sensitivity to ancillary spin configurations and inter-spin interactions. Notably, the Ising model exhibited a significant dependence on the presence of bonds between ancillary spins to achieve extractable work. This behavior underscores the cooperative nature of energy transfer within the Ising model's framework.

When analyzing the ergotropy and state fidelity outcomes, we observed an intriguing relationship between ancillary spin connections and extractable work. Significant amount of extractable work was observed when there existed a bond between ancillary spins. This phenomenon suggests that inter-spin bonds play a pivotal role in facilitating energy flow and storage within the Ising model. Such sensitivity to bond configurations underscores the importance of cooperative interactions in energy extraction from quantum battery systems operating under the Ising interaction.

Upon closer examination of the Ising model, a particular configuration emerged as the optimal charging scenario. This configuration consisted of four ancillary spins with connections between them. Intriguingly, this setup displayed a substantial amount of extractable work, signifying its potential for efficient energy extraction. However, despite the high extractable work achieved, the fidelity of the battery system in this configuration did not approach unity.We observe that while the fidelity remains below unity, it's crucial to note that the battery system is not in its ground state after the charging process.

#Figure Caption
The ergotropy difference of different configurations for Ising Model shows correleation with the number of the ancillary spins when the charger spins are connected. The best case scenario, 4-4 figure(b), shows non unity fidelity where it is still distinct form its ground state and results with significant amount of ergotropy.
